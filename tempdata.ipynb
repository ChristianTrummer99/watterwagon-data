```
import numpy.matlib 
import numpy as np 
import math as m, pandas as pd
import matplotlib.pyplot as plt

```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df2 = pd.read_csv(io.BytesIO(uploaded['tempdata.csv']),names=["Temperature"])

Tf = np.zeros(len(df2))
Tc = np.zeros(len(df2))
timestamp = []

for i in range(0,len(df2)):
  string = df2['Temperature'][i]
  string = string.split()
  string.remove('->')
  string.remove('Temperature:')
  string.remove('F;')
  string.remove('C')
  Tf[i] = float(string[1])
  Tc[i] = float(string[2])
  time = np.arange(0,len(df2))/60.0
  timestamp.append(string[0])
  df2['Temperature'][i] = string


print('timestamp =', timestamp, '\ntime =',time,'\nTempFaren =', Tf, '\nTc = ',Tc)
```

```
charge_time = np.array([0,7,14,36,46,52,60,65,89,100,104,109,111,115,119,123])
percentage = np.array([3,10,19,42,52,60,69,73,83,88,90,90,91,92,93,94])

plt.plot(time,Tf)
plt.xlabel('$Time\\,(min)$',fontsize=20)
plt.ylabel('$Temperature\\,(Farenheit)$', fontsize=20)
plt.title('Temperature(F) of USB converter vs. Time', fontsize=20)
plt.gcf().set_size_inches(12, 8)
plt.show()

plt.plot(charge_time,percentage,'ro-')
plt.xlabel('$Time\\,(min)$',fontsize=20)
plt.ylabel('Percentage of Phone Charge (%)', fontsize=20)
plt.title('Percentage Charge vs. Time', fontsize=20)
plt.gcf().set_size_inches(12, 8)
plt.show()

```

```
plt.plot(time,Tc)
fig = plt.gcf()
fig.set_size_inches(12, 8)
plt.xlabel('$Time\\,(min)$',fontsize=20)
plt.ylabel('$Temperature\\,(Celsius)$', fontsize=20)
plt.title('Temperature(C) of USB converter vs. Time', fontsize=20)
plt.show()
```
